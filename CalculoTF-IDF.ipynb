```python
# Imports
import pandas as pd
import nltk
import spacy
import re
import string
import numpy as np
from nltk.corpus import stopwords
```

```python
#apenas para ler nossos arquivos corpus
import os
from nltk.corpus.reader.plaintext import PlaintextCorpusReader
```

```python
corpusdir = 'corpus/' 
```

```python
# lendo todos os arquivos da pasta corpus
newcorpus = PlaintextCorpusReader(corpusdir, '.*')
```

```python
corpusTwittes = PlaintextCorpusReader.sents(newcorpus)
```

```python
# Listas para receber as frases e as palavras dos posts
dados_postsTwitter_frases = []
dados_postsTwitter_palavras = []
```

```python
# Loop para a tokenização
for sentence in corpusTwittes:
    dados_postsTwitter_frases.append([word.lower() for word in sentence if word.isalpha()])
    for word in sentence:
        if word.isalpha():
            dados_postsTwitter_palavras.append(word.lower())
```

```python
# Vamos converter a lista de palavras em um conjunto (set)
dados_postsTwitter_palavras = set(dados_postsTwitter_palavras)
```

## frequencia dos termos

```python
# Função para calcular a Termo Frequência
def TermFreq(documento, palavra):
    doc_length = len(documento)
    ocorrencias = len([w for w in documento if w == palavra])
    return ocorrencias / doc_length
```

```python
# printando um twiiter (frase)
dados_postsTwitter_frases[20]
```

```python
# Aplica a função para calcualr o Termo Frequência na frase acima (indice 20) pesquisando por "lula"
TermFreq(dados_postsTwitter_frases[20], 'lula')
```

```python
# Criamos um corpus Bag of words
def cria_dict():
    output = {}
    for word in dados_postsTwitter_palavras:
        output[word] = 0
        for doc in dados_postsTwitter_frases:
            if word in doc:
                output[word] += 1
    return output
```

```python
# Cria o dicionário
df_dict = cria_dict()
```

```python
# Filtra o dicionario onde contem "lula" e mostrar a quantidade de twitter/frases que contem o nome dele
df_dict['lula']
```

## Frequência Inversa

```python
# Função para calcular a Frequência Inversa de Documentos
def InverseDocumentFrequency(word):
    N = len(dados_postsTwitter_frases)
    try:
        df = df_dict[word] + 1
    except:
        df = 1
    return np.log(N/df)
```

```python
InverseDocumentFrequency('lula')
```

### TF/IDF

```python
# Função TF-IDF
def TFIDF(doc, word):
    tf = TermFreq(doc, word)
    idf = InverseDocumentFrequency(word)
    return tf * idf
```
